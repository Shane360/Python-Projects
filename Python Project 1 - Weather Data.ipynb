# What is the Mean value of each column against each "Weather Condition"?

```python
weather_data.head()
```

```python
# Use the groupby() method
# Append the mean() to 


weather_data.groupby('Weather Condition').get_group('Fog')
```

```python
weather_data.groupby('Weather Condition').mean()
```

```python
weather_data.info()
```

```python
# When you try to get the mean value of each column against the values in the Weather Condit, it throws a TypeError because there are non-numeric columns in the df
# So you have two options in order to get the desired results
# 1. define numeric columns in a variable
# 2. define non-numeric columns in a variable
# 3. use groupby() to group the numeric columns by the Weather column whilst getting the mean of the columns

non_num_cols = [col for col in weather_data.columns if weather_data[col].dtype != np.number]
num_cols = [col for col in weather_data.columns if col not in non_num_cols]

weather_data.groupby('Weather Condition')[num_cols].mean()

```
